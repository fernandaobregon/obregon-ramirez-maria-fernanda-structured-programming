## 1. Core Concepts
What is a Dictionary?

A dictionary is a data structure that stores information as key-value pairs.

### General structure:
- key -> value
#### Example:
- "name" -> "Alice"
- "age" -> 20
- "major" -> "Computer Science"

Unlike lists, which store elements using numerical positions (indexes), dictionaries store and retrieve information using unique keys.

### Why Do We Need Dictionaries?
Consider the following example using lists:



````python
students = ["Alice", "Bob", "Charlie"]
grades = [95, 80, 100]
````

To find Charlie's grade:



````python
index = students.index("Charlie")
print(grades[index])
````



This approach becomes inefficient and difficult to maintain as the amount of data grows.

Using a dictionary:


````python

grades = {
    "Alice": 95,
    "Bob": 80,
    "Charlie": 100
}

print(grades["Charlie"])
````

The dictionary directly associates each student with a grade.



## Key Concepts

A dictionary consists of:

### Key

The identifier used to access information.

```python
"name"
```

### Value

The data associated with a key.

```python
"Alice"
```

### Key-Value Pair

```python
"name": "Alice"
```

### Dictionary:

```python
student = {
    "name": "Alice",
    "age": 20,
    "major": "Computer Science"
}
```


# Dictionary vs List

| List                 | Dictionary               |
| -------------------- | ------------------------ |
| Uses indexes         | Uses keys                |
| Ordered sequence     | Key-value mapping        |
| Access by position   | Access by identifier     |
| Example: students[0] | Example: student["name"] |

Understanding this distinction is the most important learning outcome of this lab.

# 7. Practical Application: Student Grade Registry
- Problem Statement
- Create a program that:
- Creates an empty dictionary.
- Requests information for three students.
- Stores each student's name and grade.
- Displays the completed dictionary.
## Example Input




```
Name: Alice
Grade: 95

Name: Bob
Grade: 80

Name: Charlie
Grade: 100
```




 Expected Output:

````python
{
    'Alice': 95,
    'Bob': 80,
    'Charlie': 100
}
````

## Exercise 5
Complete the program below.

````python
# Exercise 5: Student Grade Registry
# This version simulates user input so the notebook can run from start to finish
# and show all required outputs clearly.

grades = {}

sample_inputs = iter(["Alice", "95", "Bob", "80", "Charlie", "100"])

def ask(prompt):
    value = next(sample_inputs)
    print(prompt + value)
    return value

for student_number in range(3):
    name = ask("Name: ")
    grade = int(ask("Grade: "))
    grades[name] = grade
    print()

print(grades)
````

````python
# Exercise 6: Retrieve a student's grade from the dictionary
# This simulates the user searching for Charlie.

student_name = "Charlie"
print("Student:", student_name)

if student_name in grades:
    print("Grade:", grades[student_name])
else:
    print("Student not found.")
````

## 9. Challenge Activity
####Character Statistics Manager

Create a dictionary containing:


```
name
health
attack
defense
level
```

The program must:
1. Display all character information.
2. Increase the level by 1.
3. Increase the attack value by 10.
4. Display the updated information.

## Workspace

````python
# Challenge Activity: Character Statistics Manager

character = {
    "name": "Astra",
    "health": 100,
    "attack": 25,
    "defense": 15,
    "level": 1
}

print("Original character information:")
for key, value in character.items():
    print(key, ":", value)

# Update character statistics
character["level"] += 1
character["attack"] += 10

print("\nUpdated character information:")
for key, value in character.items():
    print(key, ":", value)
````

## Critical Thinking Questions
1. What is the main difference between a list and a dictionary?

Answer: A list stores values by position using numeric indexes, while a dictionary stores values using meaningful keys.

2. Why are dictionaries more suitable than lists for storing student grades associated with names?

Answer: Dictionaries are more suitable because each student name can be used as a key, making it easier and faster to find the grade that belongs to that student.

3. Give one real-world example where a dictionary would be more useful than a list.

Answer: A dictionary would be useful for a contact list where each person's name is connected to their phone number, for example: `{"Alice": "555-1234"}`.

4. What would happen if two different values were assigned to the same key?

Answer: The newest value would replace the previous value because dictionary keys must be unique.

5. In your own words, explain the concept of a key-value pair.

Answer: A key-value pair is a connection between a label and its information. The key identifies the data, and the value is the data stored for that key.
