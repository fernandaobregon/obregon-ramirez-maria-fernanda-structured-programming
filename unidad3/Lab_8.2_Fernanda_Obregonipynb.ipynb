# Part 2. What Is a Procedure?

In structured programming, there are two main types of subprograms:

## Function

A function performs a task and returns a value.

Example:



````python
def calculate_area(base, height):
    return base * height

area = calculate_area(5, 4)

print(area)
````



## Procedure

A procedure performs an action but does not return a value.

Example:



````python
def display_menu():
    print("1. Add")
    print("2. Subtract")
    print("3. Exit")

display_menu()
````

A recursive function is a function that calls itself.

Observe the following example.



````python
def countdown(n):

    print(n)

    if n > 1:
        countdown(n - 1)

countdown(5)
````



Execute the program.

### Questions

1. Which numbers are displayed?
 5,4,3,2,1

2. What happens when `n` becomes 1?
 When n becomes 1, the function prints 1 and then stops calling itself.
3. Why does the program stop?
 Because the if statement its not longer true, when n equals 1, so the recursive call is not executed.



# Part 7. Understanding the Base Case

Every recursive function needs a condition that stops the recursion.

Observe the following example.



````python
def say_hello(n):

    if n == 0:
        return

    print("Hello")
    say_hello(n - 1)

say_hello(4)
````

### Questions

1. How many times is "Hello" printed?
 4 times

2. What is the base case?
 n==0

3. What would happen if the `if` statement were removed?
 The function woul keep calling itself, because there would be no base case to stop it

# Part 9. First Recursive Mathematical Example

The factorial of a number is defined as:

```text
4! = 4 × 3 × 2 × 1
```

Observe the following code.

```python
def factorial(n):

    if n == 1:
```

# Part 10. Final Challenge

Create a program that includes:

## Procedures

* display_menu()
* say_goodbye()

## Functions

* calculate_area()
* calculate_average()

## Recursive Function

* countdown()

The menu should allow the user to select an option.

```text
1. Calculate Area
2. Calculate Average
3. Countdown
4. Exit
```

### Requirements

* Use at least two procedures.
* Use at least two functions that return values.
* Use one recursive function.
* Organize the program into subprograms.


1. What is the difference between a function and a procedure?
 A function returns a value, while a procedure performs an action but does not return a value.

2. What is recursion?
When a function recalls itself

3. Why does every recursive function need a base case?
 Every recursive function needs a base case to stop the recursion and avoid an infinite loop, it serves as a brake to stop the recursive function.

4. Which topic was most difficult for you?
 Recursive functions

````python
def display_menu():
    print("\nPlease select an option:")
    print("1. Calculate Area")
    print("2. Calculate Average")
    print("3. Countdown")
    print("4. Exit")

def say_goodbye():
    print("Goodbye!")

def calculate_area(base, height):
    return base * height

def calculate_average(num1, num2, num3):
    return (num1 + num2 + num3) / 3

def countdown(n):
    print(n)

    if n > 1:
        countdown(n - 1)


option = ""
while option != "4":
    display_menu()
    option = input("Enter your option: ")

    if option == "1":
        base = float(input("Enter the base: "))
        height = float(input("Enter the height: "))

        area = calculate_area(base, height)
        print("The area is:", area)

    elif option == "2":
        num1 = float(input("Enter the first number: "))
        num2 = float(input("Enter the second number: "))
        num3 = float(input("Enter the third number: "))

        average = calculate_average(num1, num2, num3)
        print("The average is:", average)

    elif option == "3":
        number = int(input("Enter a number for the countdown: "))
        countdown(number)

    elif option == "4":
        say_goodbye()

    else:
        print("Invalid option. Please try again.")
````
