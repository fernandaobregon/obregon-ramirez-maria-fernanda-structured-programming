# Problem Description

Describe the problem you will solve:

```text
A system that calculates the average grade of students and determines if they pass or fail.

# 1. 1. Imperative Programming  

## Brief Explanation
Imperative programming works by giving the computer a series of step-by-step instructions to follow in a specific order. In this example, we use a loop to go through the grades, calculate the total sum, and then divide it by the number of grades to find the average.

## Code Example
Complete the steps

````python
# Given data
grades = [70, 80, 90]

# 1. Calculate sum using a loop
total = 0

for grade in grades:
    total += grade

# 2. Compute average
average = total / len(grades)

# 4. Evaluate if the student pass or fail
if average >= 70:
    print("Pass")
else:
    print("Fail")

# 5. Print result
print("Average:", average)
````

## Your Modification
What did you change and why?
I added a loop to go through the entire array, adding all the values inside it, then divided the total by the number of items in the array to calculate the average, and finally printed the results.


## Reflection
What did you learn from this approach and how does it compare to others?
I learned that imperative programming requires giving clear instructions step by step. It is easy to understand because the program follows a sequence of commands. Compared to other approaches, like functional programming, imperative programming gives more control over how tasks are completed but may require more lines of code.

# 1. 2. Structured Programming  

## Brief Explanation
For Structured Programming, the idea is to organize the code into clear sections or functions instead of putting everything together step by step in one block.


## Code Example
Complete the steps

````python
# Given data
grades = [70, 80, 90]

# Function to calculate average
def calculate_average(grades):
    total = 0

    for grade in grades:
        total += grade

    average = total / len(grades)
    return average


# Evaluate if the student pass or fail
if average >= 70:
    print("Pass")
else:
    print("Fail")

# Main program
result = calculate_average(grades)

# Print result
print("Average:", result)
````

## Your Modification
What did you change and why?
I added a function to calculate the average based on my grades variable, using the same loop and logic as before. Then, I returned the average value, stored the result in a variable called result, and finally printed the result.


## Reflection
I learned that structured programming makes code easier to organize and understand by separating tasks into functions. Compared to imperative programming, it is more organized and reusable, especially for larger programs.

# 1. 3. Modular Programming  

## Brief Explanation
For Modular Programming, the key idea is dividing the program into independent modules/functions, where each part has a specific responsibility. So instead of one function doing everything, you separate tasks.

## Code Example
Complete the steps

````python
# Given data
grades = [70, 80, 90]

# Module 1: Calculate average
def calculate_average(grades):
    total = 0

    for grade in grades:
        total += grade

    return total / len(grades)

# Module 2: Evaluate pass or fail
def evaluate_student(average):
    if average >= 70:
        return "Pass"
    else:
        return "Fail"

# Module 3: Print results
def print_results(average, status):
    print("Average:", average)
    print(status)

# Main program
average = calculate_average(grades)
status = evaluate_student(average)

print_results(average, status)
````

## Your Modification
What did you change and why?
I divided the program into different functions, where each function has a specific task. One function calculates the average using the grades variable, another evaluates if the student passed or failed, and another prints the results. This makes the code more organized and easier to reuse.


## Reflection
I learned that modular programming makes code easier to organize by separating tasks into different functions. It helps make the program easier to understand and reuse.

# 1. 4. Object-Oriented Programming (OOP)

## Brief Explanation
the idea is to create a class that represents the student and stores both the data (grades) and the actions (calculate average, evaluate pass/fail, print results).

## Code Example
Complete the steps

````python
# Class
class Student:

    # Save grades
    def __init__(self, grades):
        self.grades = grades

    # Calculate average
    def calculate_average(self):
        return sum(self.grades) / len(self.grades)

    # Check if student pass or fail
    def evaluate_student(self):
        if self.calculate_average() >= 70:
            return "Pass"
        else:
            return "Fail"


# Create student object
student = Student([70, 80, 90])

# Print results
print("Average:", student.calculate_average())
print(student.evaluate_student())
````

## Your Modification
What did you change and why?
I used Object-Oriented Programming by creating a class called Student. The class stores the grades and has methods to calculate the average and evaluate if the student passed or failed. This makes the code more organized because everything related to the student is grouped together.


## Reflection
I learned that Object-Oriented Programming organizes code using classes and objects. It helps group data and functions together, making the code easier to organize and reuse. At first it seems more complex, but it can be useful for bigger programs.

# 1. 5. Functional Programming (basic level)

## Brief Explanation
The idea is to use functions and avoid changing variables too much. At a beginner level, teachers usually expect you to use functions like sum() and maybe map() or lambda, but keep it simple.

## Code Example
Complete the steps

````python
# Given data
grades = [70, 80, 90]

# Function to calculate average
def calculate_average(grades):
    return sum(grades) / len(grades)

# Function to evaluate pass or fail
def evaluate_student(average):
    return "Pass" if average >= 70 else "Fail"

# Main program
average = calculate_average(grades)
status = evaluate_student(average)

# Print results
print("Average:", average)
print(status)
````

## Your Modification
What did you change and why?
I used Functional Programming by creating functions to calculate the average and evaluate if the student passed or failed. Instead of using many variables and loops, I used built-in functions like sum() to make the code simpler and easier to read.


## Reflection
Functional Programming focuses on using functions to solve problems. It can make code shorter and easier to understand because it avoids unnecessary steps. Compared to other paradigms, it feels simpler for solving small problems.
